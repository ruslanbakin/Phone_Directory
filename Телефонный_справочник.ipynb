```python
class Worker:
    def __init__(self, name):
        self.name = name

class Room:
    def __init__(self, number_room):
        self.number_room = number_room

class PhoneNumber:
    def __init__(self, number_phone, room_number=None, *worker_names):
        self.number_phone = number_phone
        self.room = Room(room_number) if room_number else None
        self.workers = [Worker(name) for name in worker_names]

    def get_full_info(self):
        worker_names = [worker.name for worker in self.workers]
        workers_info = ", ".join(worker_names) if worker_names else "работников нет"
        room_number = self.room.number_room if self.room else "Нет помещения"
        return f"Номер телефона: {self.number_phone}, Номер помещения: {room_number}, Рабоники: {workers_info}"

    def __str__(self):
        return self.get_full_info()



class PhoneDirectory:
    def __init__(self,file_name='phone_directory.txt'):
        self.phone_numbers = []
        self.file_name = file_name

    def display_directory(self):
      try:
          with open(self.file_name, 'r', encoding='utf-8') as file:
              entries = file.readlines()

              if not entries:
                  print("Справочник пуст.")
                  return
              for index, entry in enumerate(entries, start=1):
                  print(f"{index}. {entry.strip()}")
      except FileNotFoundError:
          print(f"Файл '{self.file_name}' не найден.")


class PhoneDirectoryManager:
    def __init__(self, file_name='phone_directory.txt'):
        self.phone_directory = PhoneDirectory()
        self.file_name = file_name

    def create_file(self):
        with open(self.file_name, 'w', encoding='utf-8'):
            pass

    def add_record(self, number, room, *workers):
        phone_number = PhoneNumber(number, room, *workers)
        self.phone_directory.phone_numbers.append(phone_number)

    def display_directory(self):
        self.phone_directory.display_directory()

    def write_to_file(self):
        with open(self.file_name, 'w', encoding='utf-8') as file:
            for phone_number in self.phone_directory.phone_numbers:
                file.write(phone_number.get_full_info() + '\n')

    def delete_from_file(self, number_phone):
        entries = []
        found = False

        with open(self.file_name, 'r', encoding='utf-8') as file:
            entries = file.readlines()

        # Создаем новый список записей, исключая ту, которую мы хотим удалить
        new_entries = []
        for entry in entries:
            entry_data = entry.strip().split(', ')
            entry_number = entry_data[0].split(': ')[1]
            if entry_number == number_phone:
                found = True
            else:
                new_entries.append(entry)

        if not found:
            print(f"Запись с номером телефона {number_phone} не найдена в файле.")
            return

        # Перезаписываем файл новым списком записей
        with open(self.file_name, 'w', encoding='utf-8') as file:
            file.writelines(new_entries)

        print(f"Запись с номером телефона {number_phone} успешно удалена из файла.")

    def save_to_file(self):
        with open(self.file_name, 'a', encoding='utf-8') as file:
            for phone_number in self.phone_directory.phone_numbers:
                file.write(phone_number.get_full_info() + '\n')

    def get_info_by_number(self, number):
        found = False
        with open(self.file_name, 'r', encoding='utf-8') as file:
            entries = file.readlines()
            for entry in entries:
                entry_data = entry.strip().split(', ')
                entry_number = entry_data[0].split(': ')[1]
                if entry_number == number:
                    room_number = entry_data[1].split(': ')[1]
                    workers = entry_data[2].split(': ')[1].split(', ')
                    print(f"Номер телефона: {number}, Номер помещения: {room_number}, Работники: {', '.join(workers)}")
                    found = True
                    break

        if not found:
            print(f"Номер телефона {number} не найден в каталоге.")

    def get_phone_by_room_number(self, room_number):
        found = False
        with open(self.file_name, 'r', encoding='utf-8') as file:
            entries = file.readlines()
            for entry in entries:
                entry_data = entry.strip().split(', ')
                current_room_number = entry_data[1].split(': ')[1]
                if current_room_number == room_number:
                    phone_number = entry_data[0].split(': ')[1]
                    print(f"Номер помещения: {room_number}, Номер телефона: {phone_number}")
                    found = True

        if not found:
            print(f"Номер помещения {room_number} не найден в каталоге.")

    def get_info_by_name(self, name):
        found = False
        with open(self.file_name, 'r', encoding='utf-8') as file:
            entries = file.readlines()
            for entry in entries:
                entry_data = entry.strip().split(', ')
                workers = entry_data[2].split(': ')[1].split(', ')
                if name in workers:
                    phone_number = entry_data[0].split(': ')[1]
                    room_number = entry_data[1].split(': ')[1]
                    print(f"Имя: {name}, Номер телефона: {phone_number}, Номер помещения: {room_number}")
                    found = True

        if not found:
            print(f"Сотрудник {name} не найден в каталоге.")



def main_menu(phone_manager):
        print("\nМеню:")
        print("add - Добавить запись в справочник")
        print("remove - Удалить запись из справочника")
        print("show - Просмотреть все записи в справочнике")
        print("save - Сохранить в файл")
        print("s_phone - Поиск по номеру телефона")
        print("s_room - Поиск по номеру помещения")
        print("s_name - Поиск по имени")
        print("exit - Выйти")
        while True:
            choice = input("Введите команду: ")

            if choice == 'add':
                number = input("Введите номер телефона: ")
                room = input("Введите номер помещения: ")
                workers = []
                for i in range(4):
                    worker_name = input(f"Введите имя сотрудника {i + 1} (или Enter для завершения): ")
                    if not worker_name:
                        break
                    workers.append(worker_name)
                phone_manager.add_record(number, room, *workers)

            elif choice == 'remove':
                print("Список:")
                phone_manager.display_directory()
                number_phone = input("Введите номер телефона для удаления из справочника: ")
                phone_manager.delete_from_file(number_phone)


            elif choice == 'show':
                phone_manager.display_directory()

            elif choice == 'save':
                phone_manager.save_to_file()
                print("Данные успешно сохранены в файл.")

            elif choice == 's_phone':
                number = input("Введите номер телефона для поиска информации: ")
                phone_manager.get_info_by_number(number)

            elif choice == 's_room':
                room_number = input("Введите номер помещения для поиска информации: ")
                phone_manager.get_phone_by_room_number(room_number)

            elif choice == 's_name':
                name = input("Введите имя работника для поиска информации: ")
                phone_manager.get_info_by_name(name)

            elif choice == 'exit':
                print("Выход из программы.")
                break


            else:
                print("Некорректный ввод. Пожалуйста, повторите попытку")


# Пример использования
if __name__ == "__main__":
    manager = PhoneDirectoryManager()
    main_menu(manager)
```
